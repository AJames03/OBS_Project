# Livestreaming Clock
---

**Created by:** Aron James L. Betinol

**Pre-requisite**
- Visual Studio Code
- HTML, CSS, JAVASCRIPT

***Description:***
- The purpose of this *Livestreaming Clock* is to provide a real-time clock for OBS Studio

---

## OBS Layout and Step by Step
---

**Steps 1:**
- Go to add, and click *Add Source*
- Find the *Browser* from the *Add Source*

**Step 2:**
- When you successfully add the Browser, double click of it.
- When the *Properties for Browser* appear click the ***Local File***

**Step 3:**
- Find the file/folder and select the *clock.html*

**Step 4:**
- Width: 1500
- Height: 700

**Step 5:**
- Then click the *Refresh* button
